# **Matrices**

El trabajo con matrices en python es simple. 

```python
a = np.array([[1,2,3],
              [1,2,3],
              [6,5,4]])
print(a)
print(type(a))
```

```python
T = a.T
print(T)
print(type(T))
```

```python
T = np.array([[1,2,3]])
print(T)
```

```python
dim = len(T)
print(dim) 
```

```python
dim_2 = np.shape(T)
dim_m, dim_n = np.shape(T)
print(dim_2)
#print(dim_m)
#print(dim_n)

m = dim_2[0]
print(m)
```

```python
zeros = np.zeros((10,2))
print(zeros)
print(np.shape(zeros))
print(np.shape(zeros.T))
```

```python
ones = np.ones((15,15))
print(ones)
```

```python
I = np.identity(15)
print(I)
print(np.shape(I))
```

```python
suma = ones + I
print(suma)
print(np.shape(suma))
```

```python
resta = ones - I
print(resta)
print(np.shape(resta))
```

```python
escalar_multi = 20 * ones
print(escalar_multi)
print(np.shape(escalar_multi))
```

```python
multiplicacion = ones * I
print(multiplicacion)
print(np.shape(multiplicacion))
```

```python
multiplicacion_ = np.dot(ones, I)
print(multiplicacion_)
print(np.shape(multiplicacion_))
```

```python
determinante = np.linalg.det(multiplicacion_)
print(determinante)
```

```python
inversa = np.linalg.inv(multiplicacion_)
print(inversa)
print(np.shape(inversa))
```

```python
inversa_ = np.linalg.inv(np.array([[12,2,3],[5,8,9],[4,5,6]]))
print(inversa_) 
```

```python
inversa_ = np.linalg.inv([[12,2,3],[5,8,9],[4,5,6]])
print(inversa_) 
```
