# Evaluar modelos

Obtener los dataframes como array de pandas

```python
X = np.asarray(dataComplete.data)
y = np.asarray(dataComplete.target)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
SVM = SVC(gamma=100)
SVM.fit(X_train, y_train)

y_pred = SVM.predict(X_test)

acc = accuracy_score(y_test, y_pred)
```

```python
n_splits = 5

lista = []
for i in [0.001,0.01,0.1,1,10,100]:
  scores = np.empty((4, n_splits))
  base_classifier = SVC(gamma=i)#, , random_state=42)

  skfolds = StratifiedKFold(n_splits=n_splits)
  for fold, (train_index, test_index) in enumerate(skfolds.split(X, y)):
      X_train = X[train_index]
      y_train = y[train_index]
      X_test = X[test_index]
      y_test = y[test_index]

      base_classifier.fit(X_train, y_train)

      y_pred = base_classifier.predict(X_test)
      scores[0, fold] = accuracy_score(y_test, y_pred)
      scores[1, fold] = f1_score(y_test, y_pred)
      scores[2, fold] = recall_score(y_test, y_pred)
      scores[3, fold] = precision_score(y_test, y_pred)
      lista.append(scores)
```
