```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Cancer_Data.csv')
data.head()
features = data.drop(['id', 'Unnamed: 32', 'diagnosis'], axis = 1)
labels = data['diagnosis']
label_encoder = LabelEncoder()
labels = label_encoder.fit_transform(labels)

features = np.asarray(features)

kmeans = KMeans(n_clusters= 10)
label = kmeans.fit_predict(features)
```

```python
filtered_label0 = features[label == 0]
plt.scatter(filtered_label0[:,0] , filtered_label0[:,1])
plt.show()
```

```python
#filter rows of original data
filtered_label2 = features[label == 2]

filtered_label8 = features[label == 8]

#Plotting the results
plt.scatter(filtered_label2[:,0] , filtered_label2[:,1] , color = 'red')
plt.scatter(filtered_label8[:,0] , filtered_label8[:,1] , color = 'black')
plt.show()
```

```python
u_labels = np.unique(label)

for i in u_labels:
    plt.scatter(features[label == i , 0] , features[label == i , 1] , label = i)
plt.legend()
plt.show()
```

```python
#Getting the Centroids
centroids = kmeans.cluster_centers_
u_labels = np.unique(label)

#plotting the results:

for i in u_labels:
    plt.scatter(features[label == i , 0] , features[label == i , 1] , label = i)
plt.scatter(centroids[:,0] , centroids[:,1] , s = 80)
plt.legend()
plt.show()
```
